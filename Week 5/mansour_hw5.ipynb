Task 1

Ho: average age = 30 (The average age of community center members is exactly 30 years old)
Ha: average age != 30 (The average age of community center members is not exactly 30 years old)

```python
#Task 2

import numpy as np
from statsmodels.stats.weightstats import ztest

# 1. Sample Data (Now using a larger sample for Z-test validity)
# Note: Z-tests are more reliable with n > 30
ages = np.array([28, 32, 35, 29, 31, 34, 30, 27, 33, 31, 
                 29, 30, 32, 28, 30, 31, 34, 29, 30, 32,
                 33, 27, 28, 31, 30, 32, 29, 30, 34, 31])

# 2. Set the Population Mean to test against (Null Hypothesis)
pop_mean_null = 30

# 3. Perform One-Sample Z-Test
# value: the mean under the null hypothesis
z_stat, p_val = ztest(ages, value=pop_mean_null)

# 4. Print Results
print(f"Z-Statistic: {z_stat:.4f}")
print(f"P-Value: {p_val:.4f}")

# 5. Decision Logic
alpha = 0.05
if p_val < alpha:
    print(f"Decision: Reject H0 (p < {alpha})")
else:
   print(f"Decision: Fail to Reject H0 (p >= {alpha})") 
```

```python
#Task 3

ages = np.array([28, 32, 35, 29, 31, 34, 30, 27, 33, 31, 
                 29, 30, 32, 28, 30, 31, 34, 29, 30, 32,
                 33, 27, 28, 31, 30, 32, 29, 30, 34, 31])

ages = np.append(ages, 110)
# 2. Set the Population Mean to test against (Null Hypothesis)
pop_mean_null = 30

# 3. Perform One-Sample Z-Test
# value: the mean under the null hypothesis
z_stat, p_val = ztest(ages, value=pop_mean_null)

# 4. Print Results
print(f"Z-Statistic: {z_stat:.4f}")
print(f"P-Value: {p_val:.4f}")

# 5. Decision Logic
alpha = 0.05
if p_val < alpha:
    print(f"Decision: Reject H0 (p < {alpha})")
else:
   print(f"Decision: Fail to Reject H0 (p >= {alpha})") 
```

Task 3: Adding this data point decreases our Z-score and increases our P-value, which means that adding a severe outlier makes our Z-test less conclusive. In other words, adding a major outlier makes it more likely that we will not be able to reject our null hypothesis because our findings will be less significant.

Task 4: If we kept the same sample mean but quadrupled our sample size, our Z-statistic would become much higher and our p-value would become much lower, meaning that our findings would become more significant and we would be more likely to reject the null hypothesis. A larger sample size makes it 'easier' to reject the null hypothesis because it reduces the abnormalcy of outliers/values that are not close to the mean. With more values, these values become more normal, and the resulting deviation from the mean becomes more representative of the actual population. Thus, our findings with the bigger sample size show us that the null hypothesis is less representative of the entire population, allowing us to reject it (if the p value is under .05).
