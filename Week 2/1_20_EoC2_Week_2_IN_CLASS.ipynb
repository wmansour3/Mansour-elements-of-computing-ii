### Section 1: Defining a Base Class
- **Objective**: Learn to create a base class to represent shared traits of all RPG characters.
- **Concepts**: Classes are blueprints for creating objects. Attributes store data about an object (e.g., name, level, health). Methods define behaviors for objects (e.g., `display_info`).

```python
print(type("Hello, world!"))
help(str)
```

```python
# Defining a Base Class: Character
class Character:
    """
    Represents a generic RPG character.

    Attributes:
        name (str): The character's name
        level (int): The character's level
        health (int): The character's current health
    """
    # def __init(self) default template for custom classes
    def __init__(self, name, level, health):
        '''
        Initializes a new Character object.

        Args:
            name (str): The character's name
            level (int): The character's level
            health (int): The character's current health
        '''
        self.name = name
        self.level = level
        self.health = health

    # Create a method that displays the character's information
    def display_info(self):
        '''
        Returns:
            str: A formatted string with the character's name, level, and current health
        '''
        return f"{self.name} (Level {self.level}) - {self.health} HP"
```

```python
# Create a base character
shamrock = Character('Shamrock', 5, 100)
# Display info with method
shamrock.display_info() # will not appear if it's not the last thing in the cell; to make it appear when it's not the last thing, print it
# Retrieving individual attributes
help(shamrock)
```

### Section 2: Creating a Subclass
- **Objective**: Extend the base class by creating a subclass with unique traits.
- **Concepts**: Subclasses inherit attributes and methods from the parent class. `super()` lets you call the parent class's methods.
- **Usage**: `Warrior` adds a `strength` attribute.

```python
# Subclass Example: Warrior
class Warrior(Character):

    #default template
    def __init__(self, name, level, health, strength):
        #super() calls the init attributes from Character
        super().__init__(name, level, health)
        self.strength = strength

    # Need new display_info() method if you want to show strength
```

### Section 3: Polymorphism
- **Objective**: Add polymorphic behavior by creating a subclass with unique methods.
- **Concepts**: Polymorphism allows shared methods to behave differently for each subclass.
- **Usage**: `Mage` adds a `mana` attribute and a `cast_spell` method.

```python
# Polymorphism in Action
class Mage(Character):
    def __init__(self, name, level, health, mana):
        super().__init__(name, level, health)
        self.mana = mana

    def cast_spell(self, mana_cost):
        '''
        Simulates the mage casting a spell

        Returns:
            str: A message indicating the spellcasting action
            mana: Update self.mana
        '''
        # Initialize the mage's specific attribute: mana
        self.mana -= mana_cost

        return f"{self.name} casts a spell using {mana_cost} mana! They now have {self.mana} mana."
    # Right cast_spell method that uses mana_cost and updates self.mana
```

### Section 4: Adding Methods for Specialized Behavior
- **Objective**: Extend the base class with new methods that allow interaction between objects.
- **Concepts**: Create a `Healer` class with a `heal` method that modifies another object's state.
- **Usage**: Demonstrates how objects can interact with one another in meaningful ways.

```python
# Adding Methods for Specialized Behavior
class Healer(Character):

    # Initialize character attributes
    def __init__(self, name, level, health, healing_power):
        
        # Inherit name, level, health attributes
        super().__init__(name, level, health)
        # Initialize the healer's specific attribute: healing_power
        self.healing_power = healing_power
    # Create healing method that interacts with other character
    def heal(self, target):
        '''
        Heals the target character.

        Args:
            self (Healer): Use the healing_power attribute as healing points.
            target (Character): The character to heal

        Returns:
            str: A message indicating the healing action
        '''
        # Increase the target's health by the healer's healing power
        target.health += self.healing_power
        # Return a message describing the healing action
        return f"{self.name} heals {target.name} for {self.healing_power} HP!"
```

### Section 5: Coding Challenge
- **Objective**: Create a new class and test it in the battle simulation.
- **Task**:
  1. Create a `Rogue` subclass of `Warrior`.
  2. Add a `sneak_attack` method that uses a stealth attack that reduces health from a `target` at `self.stealth * 3`.

```python
# Coding Challenge: Add Your Own Class
class Rogue(Warrior):
    # Initialize warrior attributes
    def __init__(self, name, level, health, strength, stealth):
        # Inherit attributes
        super().__init__(name, level, health, strength)
        # Initialize the Rogue's specific attribute: stealth
        self.stealth = stealth
    # Create sneak attack that targets and reduces health of another player
    def sneak_attack(self, target):
        # Reduce the target's health by a factor of stealth
        target.health -= self.stealth * 3
        # Return a message describing the sneak attack action
        return f"{self.name} stealth attacks {target.name} and lowers their HP to {target.health}"
```

### Section 6: Simulating Dynamic Behavior
- **Objective**: Use a single function to handle objects of multiple types dynamically.
- **Concepts**: Demonstrate how `isinstance` checks enable dynamic behavior.
- **Usage**: Simulate battle actions for different character types.

```python
# Simulating Dynamic Behavior
def battle_round(character, target):

    # Display the character's information
    print("\nBefore round info:")
    print(character.display_info())
    print(target.display_info())

    print("\nAction taken:")
    # Check the character's type and perform the corresponding action
    if isinstance(character, Rogue):
        print(character.sneak_attack(target))
        # If the character is a Rogue
            # Perform a sneak attack
    elif isinstance(character, Mage):
        print(character.cast_spell())# If the character is a Mage
            # Cast a spell
    elif isinstance(character, Healer):
        print(character.heal(target))
        # If the character is a Healer
            # Heal the target
    elif isinstance(character, Warrior):
        print(character.display_info())
        # If the character is a Warrior
            # Display the warrior's strength
    else:
        print("Not a valid character.")

    # Display the character's information
    print("\nAfter round info:")
    print(character.display_info())
    print(target.display_info())
```
