To compare Simple Linear Regression (SLR) and Multiple Linear Regression (MLR), we'll use the `statsmodels` library, which provides detailed statistical summaries (like P-values and R-squared).

In this example, we predict __House Price__ based on __Square Footage__ (SLR) and then add __Age__ (MLR).

### Key Components:
Dependent Variable (y): The outcome to predict (House Price).

Independent Variable (y): The predictor (Square Footage and Age).

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# 1. Create a dummy dataset
data = {
    'SqFt': [1500, 1800, 2400, 3000, 3500, 4000],
    'Age': [10, 15, 5, 20, 2, 8],
    'Price': [250000, 280000, 400000, 410000, 550000, 590000]
}
df = pd.DataFrame(data)

# --- SIMPLE LINEAR REGRESSION (SLR) ---
# Predict Price using only SqFt
X_simple = sm.add_constant(df['SqFt']) # Add intercept
model_slr = sm.OLS(df['Price'], X_simple).fit()

# --- MULTIPLE LINEAR REGRESSION (MLR) ---
# Predict Price using SqFt AND Age
X_multi = sm.add_constant(df[['SqFt', 'Age']]) # Add intercept
model_mlr = sm.OLS(df['Price'], X_multi).fit()

# 2. Compare Results
print(f"SLR R-squared: {model_slr.rsquared:.3f}")
print(f"MLR R-squared: {model_mlr.rsquared:.3f}")

print("\n--- MLR Coefficients ---")
print(model_mlr.params)
```

### Understanding the Output

__R-squared:__ The "MLR R-squared" is higher than the SLR. This means that adding "Age" helped explain more of the variance in the house prices.

__The Coefficients (B):__ In the MLR output, a coefficient for `SqFt` and a negative coefficient for `Age`. This tells you that as `Age` increases, `Price` tends to decrease, even if `SqFt` stays the same.

This is because the coefficient shows the change in (y = `SqFt`,`Age`) when there is an increase in (x=`Price`).

__The Constant:__ Both models include a const (B_o), which represents the baseline price before accounting for the other factors. 

The constant is known as the __intercept__ or__(B_o_)__ and represents the predicted (y) when (x) equals zero.

### Logistic Regression

Let's use the example above and switch from needing a Linear model to a Logistic model. 
To do this, we need to change the target variable `Price` into a binary outcome (e.g., Sold, where 1 = Yes, 0 = No) and use the `sm.Logit` function.

Here's the updated code:


To begin, make sure you are in your __eoc__ environment (terminal > `conda activate eoc`).

Now, install a new library we will be using today: `conda install scikit-learn`, then `y`

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# 1. Create a dummy dataset for classification
# Target: 'Sold' (1 if the house sold within a week, 0 if not)
data = {
    'SqFt': [1500, 1800, 2400, 3000, 3500, 4000],
    'Age': [10, 15, 5, 20, 2, 8],
    'Sold': [0, 0, 1, 0, 1, 1] 
}
df = pd.DataFrame(data)

# 2. --- SIMPLE LOGISTIC REGRESSION ---
# Predict 'Sold' using only SqFt
X_simple = sm.add_constant(df['SqFt'])
model_simple_logit = sm.Logit(df['Sold'], X_simple).fit(disp=0)

# 3. --- MULTIPLE LOGISTIC REGRESSION ---
# Predict 'Sold' using SqFt AND Age
X_multi = sm.add_constant(df[['SqFt', 'Age']])
model_multi_logit = sm.Logit(df['Sold'], X_multi).fit(disp=0)

# 4. Compare Results (Using Pseudo R-squared for Logistic)
print(f"Simple Logit Pseudo R-squared: {model_simple_logit.prsquared:.3f}")
print(f"Multiple Logit Pseudo R-squared: {model_multi_logit.prsquared:.3f}")
# 5.
print("\n--- Multiple Logistic Coefficients ---")
print(model_multi_logit.params)
```

### Understanding the Code

__Step 1:__ In Logistic Regression, the target must be binary (0 or 1). 
Here, we are modeling the probability of a house being "Sold."

__Step 2:__ Statsmodels does not include an intercept (the "b" in y=mx+b) by default. 
This line adds a column of 1s to your data so the model can calculate the baseline probability when all other variables are zero.

__Step 3:__ 
* `sm.Logit` tells the library to use the Logistic function (the S-curve) instead of a straight line.
* `.fit()` runs the optimization algorithm to find the best coefficients.
* `disp=0` tells Python to run "silently" without printing the technical iteration details.

__Step 4:__ Pseudo R-squared: Unlike linear regression, logistic regression doesn't have a perfect "R-squared." We use McFadden’s Pseudo R-squared. It ranges from 0 to 1, where values between 0.2 and 0.4 typically indicate an excellent model fit.

__Step 5:__ In OLS, the coefficients tell you the direct change in price. 
In Logistic Regression, these coefficients represent the Log-Odds.

* Positive coefficient: As this variable increases, the probability of the outcome being "1" (Sold) increases.

* Negative coefficient: As this variable increases, the probability of the outcome being "1" decreases.



## Polynomial Regression

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# 1. Create a dummy dataset
data = {
    'SqFt': [1500, 1800, 2400, 3000, 3500, 4000],
    'Age': [10, 15, 5, 20, 2, 8],
    'Price': [250000, 280000, 400000, 410000, 550000, 590000]
}
df = pd.DataFrame(data)

# --- POLYNOMIAL FEATURE ENGINEERING ---
# Create a squared term for SqFt to capture non-linear price growth
df['SqFt_Squared'] = df['SqFt'] ** 2

# --- POLYNOMIAL REGRESSION (Degree 2) ---
# Predict Price using SqFt AND SqFt^2
X_poly = sm.add_constant(df[['SqFt', 'SqFt_Squared']])
model_poly = sm.OLS(df['Price'], X_poly).fit()

# --- MULTIPLE POLYNOMIAL REGRESSION ---
# Adding Age back into the mix
X_poly_multi = sm.add_constant(df[['SqFt', 'SqFt_Squared', 'Age']])
model_poly_multi = sm.OLS(df['Price'], X_poly_multi).fit()

# 2. Compare Results
print(f"Polynomial R-squared: {model_poly.rsquared:.3f}")
print(f"Poly + Age R-squared: {model_poly_multi.rsquared:.3f}")

print("\n--- Polynomial Coefficients ---")
# The 'SqFt_Squared' coefficient explains the "curvature" of the price trend
print(model_poly_multi.params)
```

### Understanding the Code

__Feature Engineering:__ Polynomial regression is essentially linear regression with "powered" versions of your features. By manually creating `df['SqFt_Squared']`, you allow the model to fit a parabola instead of just a straight line.

__The Model:__ We still use `sm.OLS.` Even though the relationship between (x) and (y) is now curved, the relationship between the coefficients and (y) remains linear, so "Ordinary Least Squares" is still the correct tool.

__Interpretation:__ A positive coefficient for the squared term indicates that price increases at an accelerating rate as square footage grows; a negative coefficient would suggest growth that "levels off".

### Understanding the Results

The outcome of `model_poly.params` tells you the "story" of the market:

__Accelerating Growth:__ If SqFt is positive and SqFt_Squared is positive, the model shows that larger homes are disproportionately more expensive (the "mansion effect").

__Diminishing Returns:__ If SqFt is positive but SqFt_Squared is negative, the model shows that while bigger is better, each extra square foot adds less value than the one before it.

__Summary: "The Verdict"__

The outcome of this code is a more flexible prediction. Instead of assuming a fixed price-per-foot, the model acknowledges that the relationship between size and price changes as houses move from "starter homes" to "estates."
