```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# 1. Create a dummy dataset
data = {
    'Distance': [3, 6, 9, 12],
    'Success': [1, 1, 0.25, 0.25],
    'Failure': [0, 0, 0.75, 0.75]
}
df = pd.DataFrame(data)

# --- SIMPLE LINEAR REGRESSION (SLR) ---
# Predict Price using only SqFt
X_simple = sm.add_constant(df['Distance']) # Add intercept
model_slr = sm.OLS(df['Failure'], X_simple).fit()

# --- MULTIPLE LINEAR REGRESSION (MLR) ---
# Predict Price using SqFt AND Age
X_multi = sm.add_constant(df[['Distance', 'Failure']]) # Add intercept
model_mlr = sm.OLS(df['Success'], X_multi).fit()

# 2. Compare Results
print(f"SLR R-squared: {model_slr.rsquared:.3f}")
print(f"MLR R-squared: {model_mlr.rsquared:.3f}")

print("\n--- MLR Coefficients ---")
print(model_mlr.params)
```
